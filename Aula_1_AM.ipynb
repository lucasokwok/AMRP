## Introdução ao aprendizado de Máquina

A **Inteligência Artificial** (IA) é uma ampla área da ciência da computação que visa desenvolver sistemas capazes de realizar tarefas que, tradicionalmente, requerem inteligência humana. Essas tarefas incluem percepção visual, reconhecimento de voz, tomada de decisão e tradução de idiomas.

A IA pode ser compreendida como um conjunto de técnicas e abordagens que são frequentemente divididas em três grandes categorias:

* IA baseada em regras (IA simbólica): Envolve sistemas baseados em regras explícitas, lógica formal e sistemas especialistas.

* IA estatística ou probabilística: Utiliza modelos matemáticos e estatísticos para lidar com incertezas, por exemplo, redes Bayesianas e inferência probabilística.

* **Aprendizado de Máquina** (*Machine Learning* - ML): Uma subárea da IA que foca no desenvolvimento de algoritmos capazes de aprender padrões diretamente dos dados, sem serem explicitamente programados para cada tarefa específica.

Dentro do aprendizado de máquina, destacam-se algumas subáreas essenciais, como:

* **Aprendizado supervisionado**: Os algoritmos aprendem a partir de dados rotulados, onde se sabe previamente o resultado esperado.

* **Aprendizado não supervisionado**: Os algoritmos encontram padrões ou agrupamentos em dados não rotulados.

* **Aprendizado semi-supervisionado**: O aprendizado semi-supervisionado é uma técnica intermediária entre o aprendizado supervisionado e o não supervisionado. Neste método, o treinamento é realizado utilizando um conjunto reduzido de dados rotulados juntamente com uma grande quantidade de dados não rotulados.

* **Aprendizado por reforço**: Os algoritmos aprendem por tentativa e erro, buscando maximizar recompensas em um ambiente interativo.

* **Aprendizado por observação/inspiração/evolução**: Os algoritmos aprendem por meio da simulação de eventos com determinada inspiração. Por exemplo, simula-se o processo de evolução natural sobre cromossomos artificiais para se solucionar um problema de otimização.

**Obs**.: Uma outra subárea, frequentemente destacada dentro do aprendizado de máquina, é o Aprendizado Profundo (*Deep Learning*), que emprega redes neurais com múltiplas camadas para aprender representações complexas diretamente dos dados brutos.

## **Aprendizado Supervisionado**: Modelos Preditivos e Estimadores em Aprendizado de Máquina


Modelos preditivos são formalmente definidos como funções matemáticas ou algoritmos que, dado um conjunto conhecido de exemplos rotulados, constroem estimadores capazes de predizer o rótulo ou valor numérico associado a novos exemplos. Estes estimadores podem ser formalizados matematicamente como funções:

$$
f: X \rightarrow Y,
$$

onde  $X$  representa o espaço dos atributos (ou variáveis independentes), e  $Y$  o espaço dos rótulos (ou variáveis dependentes). A construção desses modelos ocorre a partir de dados históricos, e o processo geral é denominado aprendizado supervisionado.

Em aprendizado supervisionado, o conjunto de treinamento é definido por:

$$
D = \{(\mathbf{x}_1, y_1), (\mathbf{x}_2, y_2), ..., (\mathbf{x}_n, y_n)\},
$$

em que $\mathbf{x}_i \in X$ é um tensor de atributos e $y_i \in Y$. Podemos supor a título de facilitação, que cada $x_i$ seja um vetor numérico de dimensão $d$ e  $y_i \in Y$ é o rótulo associado ao exemplo $i$.

Duas das principais categorias de estimadores usados no aprendizado supervisionado são **classificação** e **regressão**, as quais são definidas principalmente pelo tipo de variável dependente.

---

## Classificação

Em problemas de classificação, o objetivo é associar cada instância a um dentre diversos rótulos nominais possíveis. Matematicamente, o espaço \( Y \) é discreto, muitas das vezes finito e não ordenado:

$$
Y = \{C_1, C_2, \dots, C_k\},\quad k \geq 2,
$$

onde cada $C_j$ é uma classe nominal ou categoria específica. Exemplos práticos incluem conjuntos como $\{0,1,2\}$, $\{-1,1\}$, $\{\text{doente, saudável}\}$, $\{\text{bom pagador, mau pagador}\}$, ou $\{\text{iris setosa, iris versicolor, iris virginica}\}$.

O estimador para classificação, denominado **classificador**, busca encontrar uma função $f_C(\mathbf{x})$:

$$
f_C(\mathbf{x}) : X \rightarrow Y.
$$

Tal função atribui a cada observação $\mathbf{x}$ exatamente uma classe $C_j$.


### **Exemplo 1**: Monte modelos preditivos que classifiquem um valor inteiro dado como sendo par ou ímpar a partir dos exemplos a seguir. Indique cada componente dos modelos ($X$, $Y$ e $f$).

```python
import pandas as pd
from IPython.display import display

# Dados da tabela
dados = {
    "Valor natural": list(range(1, 11)),
    "Classe original": ["Ímpar", "Par", "Ímpar", "Par", "Ímpar", "Par", "Ímpar", "Par", "Ímpar", "Par"],
}

# Criando o DataFrame
df = pd.DataFrame(dados)

df = df.set_index("Valor natural")

# Exibir a tabela no formato do Jupyter Notebook
display(df)
```

```python
# prompt: scatter plot de x colorido

import matplotlib.pyplot as plt

def plot_df(df):

  # Criando o gráfico de dispersão
  plt.figure(figsize=(8, 6))
  plt.scatter(df.index, [1 if classe == "Ímpar" else 0 for classe in df["Classe original"]], c=["r" if classe == "Ímpar" else "g" for classe in df["Classe original"]])

  # Adicionando rótulos e título
  plt.xlabel("Valor natural")
  plt.ylabel("Classe (0 = Par, 1 = Ímpar)")
  plt.title("Gráfico de Dispersão da Classe por Valor Natural")

  # Exibindo o gráfico
  plt.show()
```

```python
plot_df(df)
```

```python
def f1(n):
  # definição lógica / hard
  if n in [1, 3, 5, 7, 9]:
    return 1
  else:
    return 0

for i in range(11,20):
  novo_df = pd.DataFrame({"Valor natural": [i], "Classe original": [f1(i)]})
  df = pd.concat([df, novo_df], ignore_index=True)

plot_df(df)
```

```python
def f2(n):
  r = n % 2
  if r == 0:
    return "Par"
  else:
    return "Ímpar"

df2 = df.copy()
for i in range(11,20):
  novo_df = pd.DataFrame({"Valor natural": [i], "Classe original": [f2(i)]})
  df2 = pd.concat([df2, novo_df], ignore_index=True)

plot_df(df2)

```

```python
import numpy as np

def f3(n):
  r = int(np.abs(np.sin(n*np.pi/2)))

  if r == 0:
    return "Par"
  else:
    return "Ímpar"

df3 = pd.DataFrame(columns=["Valor natural", "Classe original"])
for i in range(0,20):
  novo_df = pd.DataFrame({"Valor natural": [i], "Classe original": [f3(i)]})
  df3 = pd.concat([df3, novo_df], ignore_index=True)

plot_df(df3)
```

## Regressão

Na regressão, por outro lado, o objetivo é prever valores contínuos associados a cada instância. Portanto, matematicamente, o espaço $ Y $ é um conjunto contínuo, tipicamente representado por números reais:

$$
Y \subseteq \mathbb{K},
$$

sendo $\mathbb{K}$ um corpo numérico.

Exemplos típicos incluem variáveis como peso, temperatura ou vazão de água. Neste caso, o estimador é denominado **regressor** e busca-se uma função:

$$
f_R(\mathbf{x}) : X \rightarrow \mathbb{K}
$$


### **Exemplo 2**: Polinômios Interpoladores vs. Mínimos Quadrados

Considere o seguinte problema: temos uma função quase linear com pequenos ruídos aleatórios adicionados. Nosso objetivo é ajustar uma curva usando:

- **Polinômio Interpolador (Grau alto)**  
- **Polinômio por Mínimos Quadrados (Grau baixo)**

Vamos analisar na prática como cada método se comporta.

---

#### 1. Gerando os Dados Originais (Quase Linear + Ruído):

Considere uma função original próxima a \(y = x + 1\):


```python

import numpy as np
import matplotlib.pyplot as plt

# Definir pontos originais
np.random.seed(42)
x = np.linspace(0, 10, 8)
y = x + 1 + np.random.normal(0, 0.2, len(x))  # Pequeno ruído gaussiano

# Plotar dados originais
plt.scatter(x, y, label='Dados originais', color='black')
plt.plot(x, x+1, '--', label='Função original (x+1)', color='gray')
plt.legend()
plt.title("Dados Originais com Ruído Pequeno")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.show()
```



---

#### 2. Polinômio Interpolador (Grau 7)

Um polinômio interpolador com \( n=8 \) pontos terá grau \( n-1=7 \) e passará exatamente por todos os pontos.  

**Código para ajuste do polinômio interpolador:**


```python
# Ajuste polinômio interpolador grau 7
coef_interp = np.polyfit(x, y, 7)
p_interp = np.poly1d(coef_interp)

# Gerando valores para plotar o polinômio interpolador
x_dense = np.linspace(0, 10, 200)
y_interp = p_interp(x_dense)

# Plotagem
plt.scatter(x, y, label='Dados originais', color='black')
plt.plot(x_dense, y_interp, '-', color='red', label='Polinômio Interpolador (grau 7)')
plt.plot(x_dense, x_dense+1, '--', color='gray', label='Função original (x+1)')
plt.legend()
plt.title("Polinômio Interpolador (grau 7)")
plt.xlabel("x")
plt.ylabel("y")
plt.ylim([0, 12])
plt.grid(True)
plt.show()
```


**Característica do interpolador:**  
- Passa exatamente pelos pontos.
- Oscila bastante, indicando potencial overfitting.

---

#### 3. Polinômio por Mínimos Quadrados (Grau 1)

Agora, ajuste um polinômio linear (grau 1), por mínimos quadrados:

**Código para ajuste mínimos quadrados:**


```python

# Ajuste polinômio grau 1 por mínimos quadrados
coef_mq = np.polyfit(x, y, 1)
p_mq = np.poly1d(coef_mq)

# Valores para plotagem
y_mq = p_mq(x_dense)

# Plotagem
plt.scatter(x, y, label='Dados originais', color='black')
plt.plot(x_dense, y_mq, '-', color='blue', label='Polinômio Mínimos Quadrados (grau 1)')
plt.plot(x_dense, x_dense+1, '--', color='gray', label='Função original (x+1)')
plt.legend()
plt.title("Polinômio por Mínimos Quadrados (grau 1)")
plt.xlabel("x")
plt.ylabel("y")
plt.ylim([0, 12])
plt.grid(True)
plt.show()
```


**Características do ajuste por mínimos quadrados:**  
- Não passa exatamente por todos os pontos.
- Muito mais estável e generalizável.

---

#### 4. Analisando Generalização: Avaliação num novo ponto (por exemplo \( x=5.5 \)):


```python

x_novo = 5.5
y_real = x_novo + 1

y_interp_novo = p_interp(x_novo)
y_mq_novo = p_mq(x_novo)

print(f"Valor real (x+1): {y_real:.2f}")
print(f"Predição Interpolador (grau 7): {y_interp_novo:.2f}")
print(f"Predição Mínimos Quadrados (grau 1): {y_mq_novo:.2f}")
```

```python
# Reimportar bibliotecas e redefinir variáveis para gerar o gráfico corretamente
import numpy as np
import matplotlib.pyplot as plt

# Dados originais

np.random.seed(42)
x = np.linspace(0, 10, 8)
y = x + 1 + np.random.normal(0, 0.2, len(x))

# Função original
x_dense = np.linspace(-2, 12, 200)

# Polinômio interpolador grau 7
coef_interp = np.polyfit(x, y, 7)
p_interp = np.poly1d(coef_interp)
y_interp = p_interp(x_dense)

# Polinômio mínimos quadrados grau 1
coef_mq = np.polyfit(x, y, 1)
p_mq = np.poly1d(coef_mq)
y_mq = p_mq(x_dense)

# Gerar gráfico comparativo completo
plt.figure(figsize=(10, 6))

# Pontos originais
plt.scatter(x, y, label='Dados originais', color='black', s=50)

# Função original (sem ruído)
plt.plot(x_dense, x_dense+1, '--', color='gray', linewidth=2, label='Função original (x+1)')

# Polinômio Interpolador (grau 7)
plt.plot(x_dense, y_interp, '-', color='red', linewidth=2, label='Interpolador grau 7')

# Polinômio Mínimos Quadrados (grau 1)
plt.plot(x_dense, y_mq, '-', color='blue', linewidth=2, label='Mínimos Quadrados grau 1')

# Configurações adicionais
plt.legend()
plt.title("Comparação entre Curvas Aproximadas e Dados Originais")
plt.xlabel("x")
plt.ylabel("y")
plt.xlim([-10, 20])
plt.grid(True)
plt.show()
```


---

#### 5. Discussão Conceitual (didática):

- **Polinômio Interpolador (alto grau):**
  - Passa por todos os pontos observados, mas tem alta complexidade e pode produzir oscilações não desejadas (fenômeno de Runge).
  - Tem forte tendência a overfitting, prejudicando a predição em novos pontos.

- **Polinômio por Mínimos Quadrados (baixo grau):**
  - É mais estável, menos sensível a ruído.
  - Generaliza melhor para novos dados.

## Erros de modelos para aprendizado supervisionado


Em aprendizado de máquina, todo modelo preditivo busca **aproximar uma função desconhecida** que relaciona entradas $ \mathbf{x} $ com saídas $ y$. Para avaliar a qualidade dessa aproximação, utilizamos **funções de perda**, que quantificam o **erro entre a predição do modelo e o valor real**.

Essas funções de perda, quando aplicadas sobre o conjunto de dados, nos fornecem diferentes tipos de erro — como o erro local, o erro de treinamento e o erro de teste — que são fundamentais para entender a capacidade de aprendizado e generalização de um modelo.

---

## Função de Perda

A **função de perda** (ou função de custo) é uma função matemática que mede o erro de predição em cada ponto. Dados:

- Um modelo $f(\mathbf{x})$
- Um exemplo de treino $ (\mathbf{x}_i, y_i) $
- Uma função de perda $\mathcal{L}(y_i, \hat{y}_i)$

Temos:

$$
\text{Erro local} = \mathcal{L}(y_i, f(\mathbf{x}_i))
$$

Exemplos de funções de perda:

- **Erro Modular:**  
$$
\mathcal{L}(y, \hat{y}) = |y - \hat{y}|
$$

- **Erro Quadrático (para regressão):**  
$$
\mathcal{L}(y, \hat{y}) = (y - \hat{y})^2
$$

- **Perda 0-1 (para classificação):**  
$$
\mathcal{L}(y, \hat{y}) = \mathbb{1}[y \ne \hat{y}]
$$

---

## Tipos de Erro

### 1. **Erro Local**

Erro computado para uma única amostra:

$$
\varepsilon_i = |y_i - f(\mathbf{x}_i)|
$$

Indica o quanto a predição individual se afasta do valor real.

---

### 2. **Erro de Treinamento**

É o valor médio da função de perda **sobre os dados usados para treinar o modelo**. No caso do MSE:

$$
\text{Erro}_{\text{treino}} = \frac{1}{n} \sum_{i=1}^{n} \mathcal{L}(y_i, f(\mathbf{x}_i))
= \frac{1}{n} \sum_{i=1}^{n} (y_i - f(\mathbf{x}_i))^2
$$

Esse erro indica **o quão bem o modelo se ajusta ao conjunto que ele viu**.

---

### 3. **Erro de Teste**

É o valor médio da função de perda **sobre um conjunto novo de dados**, não utilizado durante o treinamento:

$$
\text{Erro}_{\text{teste}} = \mathbb{E}_{(\mathbf{x}, y) \sim \mathcal{D}}[\mathcal{L}(y, f(\mathbf{x}))]
$$

Esse erro é estimado na prática usando um **conjunto de teste** e indica a **capacidade de generalização do modelo**.

---

## Relação com Viés e Variância

A análise de **viés e variância** nos ajuda a entender a origem dos erros do modelo e a tomar decisões sobre sua complexidade.

### Viés (Bias)

É o erro introduzido por **assumir uma forma incorreta para a função $f$**.  
Modelos com **alto viés** são muito simples e **não conseguem capturar os padrões dos dados**.

$$
\text{Bias}[f(\mathbf{x})] = \mathbb{E}[f(\mathbf{x})] - f^*(\mathbf{x})
$$

(onde $f^*(\mathbf{x})$ é a função verdadeira que gera os dados)

---

### Variância (Variance)

É o erro introduzido pela **sensibilidade do modelo aos dados de treino**.  
Modelos com **alta variância** mudam muito suas predições com pequenas mudanças nos dados.

$$
\text{Variance}[f(\mathbf{x})] = \mathbb{E}\left[(f(\mathbf{x}) - \mathbb{E}[f(\mathbf{x})])^2\right]
$$

---

### Erro Esperado Total

A decomposição clássica do erro esperado (para regressão com MSE) é:

$$
\text{Erro total} = \underbrace{\text{viés}^2}_{\text{erro sistemático}} +
\underbrace{\text{variância}}_{\text{instabilidade do modelo}} +
\underbrace{\text{ruído irredutível}}_{\text{aleatoriedade dos dados}}
$$

Essa decomposição mostra que **há um limite inferior para o erro** (o ruído), e que o modelo ideal deve **equilibrar bias e variância**.

---

```python
import numpy as np
import matplotlib.pyplot as plt

# Definindo a função alvo: y = x + 1
def funcao_real(x):
    return x + 1

# Dados de entrada (3 pontos)
X = np.array([0, 1, 2])
y = funcao_real(X)
y_obs = y + np.random.normal(0, 0.2, len(y))

# Ajuste de um polinômio de grau 2 (quadrático) aos 3 pontos
coef = np.polyfit(X, y_obs, deg=2)
f_aproximador = np.poly1d(coef)

# Escolher um ponto fora da base para ilustrar o erro local
x_exemplo = 3
y_real = funcao_real(x_exemplo)
y_pred = f_aproximador(x_exemplo)
erro_local = y_real - y_pred

# Plotando a função real, a aproximação e o erro local
x_plot = np.linspace(-0.5, 3.5, 200)
y_real_plot = funcao_real(x_plot)
y_aprox_plot = f_aproximador(x_plot)

plt.figure(figsize=(10, 6))
plt.plot(x_plot, y_real_plot, '--', label='$y = x + 1$', color='gray')
plt.plot(x_plot, y_aprox_plot, label='Aproximação Quadrática', color='blue')
plt.scatter(X, y_obs, color='black', zorder=5, label='Pontos de treinamento')

# Destacar o ponto de erro
plt.scatter([x_exemplo], [y_real], color='green', label='Valor real em $x=3$', zorder=5)
plt.scatter([x_exemplo], [y_pred], color='red', label='Predição do modelo em $x=3$', zorder=5)

# Linha vertical indicando o erro local
plt.plot([x_exemplo, x_exemplo], [y_pred, y_real], 'r--', linewidth=2)
plt.annotate(f'Erro local ($|y_i - f(x_i)|$) = {np.abs(erro_local):.2f}', xy=(x_exemplo, (y_pred + y_real)/2),
             xytext=(x_exemplo + 0.2, (y_pred + y_real)/2 + 0.5),
             arrowprops=dict(arrowstyle='->', color='red'), color='red')

plt.title("Erro Local ($y_i - f(x_i)$) na Aproximação de $y = x + 1$ por um Polinômio Quadrático $f(x)$")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```


## Overfitting e Underfitting: Conceitos, Fundamentação e Exemplos

No contexto do aprendizado supervisionado, um dos principais desafios é **atingir o equilíbrio entre um modelo que seja suficientemente expressivo para aprender os padrões dos dados** e, ao mesmo tempo, **simples o bastante para generalizar bem para dados novos**.

Dois problemas clássicos surgem quando esse equilíbrio é rompido: **overfitting (sobreajuste)** e **underfitting (subajuste)**. Ambos representam falhas do modelo em capturar corretamente a estrutura dos dados.

---

##  Overfitting: Conceito e Fundamentação

###  Definição

**Overfitting**, ou **sobreajuste**, ocorre quando o modelo aprende os **dados de treinamento com tanta precisão**, que **acaba capturando também o ruído, exceções e irregularidades** específicas daquele conjunto — em vez de aprender apenas os padrões gerais.

Isso leva a um comportamento onde o modelo **desempenha muito bem no treinamento**, mas **falha ao prever dados novos**.

---

### Fundamentação Teórica

Dado um conjunto de treinamento:

$$
D = \{(\mathbf{x}_1, y_1), (\mathbf{x}_2, y_2), ..., (\mathbf{x}_n, y_n)\}
$$

Um modelo ajustado $ f $ pode ter desempenho ideal se:

$$
\text{Erro em treinamento} \approx \text{Erro em generalização}
$$

No caso de overfitting, ocorre:

$$
\text{Erro em treinamento} \ll \text{Erro em generalização}
$$

---

###  Em Regressão

Na **regressão**, o overfitting frequentemente se manifesta como uma **curva altamente oscilante**, que passa exatamente pelos pontos de dados, mesmo quando há ruído.

#### Exemplo:
Um polinômio de grau 7 ajustado a 8 pontos pode parecer excelente visualmente no conjunto original, mas falha drasticamente ao estimar valores intermediários.

---

### 📊 Em Classificação

Na **classificação**, o sobreajuste aparece quando o modelo **cria fronteiras de decisão complexas e recortadas**, que se ajustam a todos os pontos do treinamento, inclusive outliers.

---

##  Underfitting: Conceito e Fundamentação

###  Definição

**Underfitting**, ou **subajuste**, ocorre quando o modelo é **muito simples para aprender a estrutura dos dados**. Ele **não consegue representar a complexidade real do fenômeno** e, por isso, apresenta **alto erro tanto no treinamento quanto na validação/teste**.

---

### Fundamentação Teórica

Um modelo subajustado falha mesmo com os dados conhecidos, e não melhora ao ser exposto a novos dados:

$$
\text{Erro em treinamento} \approx \text{Erro em teste} \gg \text{Erro ideal}
$$

Isso geralmente ocorre quando se usa:

- Um modelo com **poucos parâmetros** (ex: uma reta para dados claramente curvos).
- Pouco tempo de treinamento.
- Poucas iterações de ajuste.

---

### Exemplo Prático

Na regressão, ajustar uma **reta** a uma função quadrática gera um modelo com **baixo poder preditivo**.

Na classificação, um modelo linear tentando separar dados que exigem uma separação não-linear simplesmente **não consegue** encontrar uma fronteira útil.

---

## Comparação Visual dos Cenários

| Cenário         | Erro em Treino | Erro em Teste | Característica                           |
|-----------------|----------------|----------------|-------------------------------------------|
| Underfitting    | Alto           | Alto           | Modelo incapaz de capturar padrões        |
| Bom ajuste      | Baixo          | Baixo          | Modelo generaliza bem                     |
| Overfitting     | Muito baixo    | Alto           | Modelo memorizou o treinamento            |

---

## Causas Comuns

| Overfitting                         | Underfitting                         |
|------------------------------------|--------------------------------------|
| Modelo muito complexo              | Modelo muito simples                 |
| Poucos dados de treinamento        | Pouco tempo de treinamento           |
| Ruído excessivo nos dados          | Poucas variáveis ou recursos         |


```python
# Gerando novamente os dados de treinamento
np.random.seed(42)
X = np.linspace(0, 10, 101)
y = 2 * X + 1 + np.random.normal(0, 2, size=X.shape)

# Ajustando modelos com grau 1 (underfit), grau 2 (ideal), grau 19 (overfit com n-1 graus)
coef_underfit = np.polyfit(X, y, deg=1)
coef_ideal = np.polyfit(X, y, deg=2)
coef_overfit = np.polyfit(X, y, deg=len(X)-1)  # grau máximo = n - 1

f_underfit = np.poly1d(coef_underfit)
f_ideal = np.poly1d(coef_ideal)
f_overfit = np.poly1d(coef_overfit)

# Escolhendo um ponto x para mostrar o erro local
x_exemplo = 4.5
y_real_exemplo = 2 * x_exemplo + 1

y_underfit_pred = f_underfit(x_exemplo)
y_ideal_pred = f_ideal(x_exemplo)
y_overfit_pred = f_overfit(x_exemplo)

# Gerando valores para o gráfico
X_plot = np.linspace(0, 10, 200)

# Plotando com destaque no erro local
plt.figure(figsize=(10, 6))
plt.scatter(X, y, color='black', label='Dados de treinamento')
plt.plot(X_plot, f_underfit(X_plot), color='red', label='Underfit (grau 1)')
plt.plot(X_plot, f_ideal(X_plot), color='green', label='Ideal (grau 2)')
plt.plot(X_plot, f_overfit(X_plot), color='blue', label=f'Overfit (grau {len(X)-1})')

# Pontos previstos no x_exemplo
plt.scatter([x_exemplo]*3, [y_underfit_pred, y_ideal_pred, y_overfit_pred],
            color=['red', 'green', 'blue'], zorder=5)

# Linhas verticais de erro
plt.plot([x_exemplo, x_exemplo], [y_real_exemplo, y_underfit_pred], 'r--', lw=2)
plt.plot([x_exemplo, x_exemplo], [y_real_exemplo, y_ideal_pred], 'g--', lw=2)
plt.plot([x_exemplo, x_exemplo], [y_real_exemplo, y_overfit_pred], 'b--', lw=2)

# Anotações com setas
plt.annotate('Erro underfit', xy=(x_exemplo, (y_real_exemplo + y_underfit_pred)/2),
             xytext=(x_exemplo + 0.5, (y_real_exemplo + y_underfit_pred)/2 + 2),
             arrowprops=dict(arrowstyle='->', color='red'), color='red')

plt.annotate('Erro ideal', xy=(x_exemplo, (y_real_exemplo + y_ideal_pred)/2),
             xytext=(x_exemplo + 0.5, (y_real_exemplo + y_ideal_pred)/2 - 2),
             arrowprops=dict(arrowstyle='->', color='green'), color='green')

plt.annotate('Erro overfit', xy=(x_exemplo, (y_real_exemplo + y_overfit_pred)/2),
             xytext=(x_exemplo + 0.5, (y_real_exemplo + y_overfit_pred)/2 + 1),
             arrowprops=dict(arrowstyle='->', color='blue'), color='blue')

# Finalização do gráfico
plt.title(f"Erro Local em Modelos com Underfit, Ajuste Ideal e Overfit (grau {len(X)-1})")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```

### Trade-off entre viés e variância

Em aprendizado de máquina, o desempenho de um modelo está diretamente relacionado ao equilíbrio entre viés e variância.
Este equilíbrio é conhecido como o 'trade-off viés-variância'. Modelos muito simples apresentam alto viés: não conseguem
capturar os padrões dos dados (underfitting). Por outro lado, modelos muito complexos têm alta variância: são instáveis
e sensíveis demais às variações do conjunto de treinamento (overfitting).

O objetivo é encontrar uma complexidade que minimize o erro total, que é composto por três partes:
• O viés ao quadrado (erro sistemático);
• A variância (erro por sensibilidade ao conjunto de dados);
• O ruído irreducível (parte aleatória dos dados que nenhum modelo pode prever).


```python
import numpy as np
import matplotlib.pyplot as plt



# Simulando o trade-off entre bias^2, variance e total error
complexidade = np.linspace(0.1, 10, 200)
bias2 = 1 / (complexidade + 1)
variancia = (complexidade**2) / (complexidade.max()**2)
erro_irredutivel = 0.1
erro_total = bias2 + variancia + erro_irredutivel

# Encontrando o ponto de mínimo da curva de erro total
indice_minimo = np.argmin(erro_total)
x_min = complexidade[indice_minimo]
y_min = erro_total[indice_minimo]

# Plotando o gráfico com anotações e áreas sombreadas
plt.figure(figsize=(12, 7))
plt.plot(complexidade, bias2, label='Viés² (Bias²)', color='red')
plt.plot(complexidade, variancia, label='Variância', color='blue')
plt.plot(complexidade, erro_total, label='Erro Total Esperado', color='black', linewidth=2)
plt.axhline(erro_irredutivel, color='gray', linestyle='--', label='Ruído irreducível')

# Sombreamento da região de underfitting e overfitting
plt.axvspan(0.1, x_min, color='red', alpha=0.05, label='Underfitting')
plt.axvspan(x_min, 10, color='blue', alpha=0.05, label='Overfitting')

# Marcando o ponto ótimo
plt.scatter([x_min], [y_min], color='black', s=100, zorder=5)
plt.annotate(f"Ponto ideal\n(Erro mínimo)", xy=(x_min, y_min),
             xytext=(x_min + 1, y_min + 0.1),
             arrowprops=dict(arrowstyle='->', color='black'),
             fontsize=10, color='black')

# Estética final
plt.title("Trade-off entre Viés² e Variância", fontsize=14)
plt.xlabel("Complexidade do Modelo", fontsize=12)
plt.ylabel("Erro", fontsize=12)
plt.legend(loc="upper left")
plt.grid(True)
plt.tight_layout()
plt.show()

```


- 🔴 **Viés²** (em vermelho): diminui à medida que o modelo se torna mais complexo, pois modelos mais expressivos conseguem se aproximar melhor da função real.
- 🔵 **Variância** (em azul): aumenta com a complexidade, pois modelos muito complexos são sensíveis às flutuações dos dados.
- ⚫ **Erro total esperado** (linha preta): soma do viés², variância e ruído — o ponto de mínimo dessa curva indica o **equilíbrio ideal**.
- Linha cinza: representa o **ruído irreducível**, ou seja, o limite inferior do erro que nenhum modelo pode ultrapassar.

## Aprendizado Não Supervisionado: Fundamentação Teórica e Aplicações

### Fundamentação Teórica

O **aprendizado não supervisionado** é uma das principais abordagens da inteligência artificial e do aprendizado de máquina. Nessa modalidade, **os algoritmos aprendem apenas a partir dos dados de entrada**, **sem rótulos ou saídas previamente definidas**.

Diferente do aprendizado supervisionado — em que os dados vêm acompanhados de uma resposta (como uma classe ou valor numérico) —, no aprendizado não supervisionado **não há “verdades” fornecidas ao algoritmo**. O sistema precisa **descobrir por conta própria padrões, relações ou estruturas** ocultas dentro dos dados.

Matematicamente, temos apenas um conjunto de dados de entrada:

$$
X = \{\mathbf{x}_1, \mathbf{x}_2, \dots, \mathbf{x}_n\}.
$$

E o objetivo do algoritmo é identificar alguma **estrutura latente**, como:

- Grupos similares de dados
- Reduções de redundância
- Tendências ou direções principais
- Valores fora do padrão (anomalias)

O aprendizado não supervisionado, portanto, é uma **abordagem exploratória**, que auxilia na **compreensão de dados complexos** e serve como base para análises posteriores, tomadas de decisão, ou mesmo como etapa de pré-processamento para algoritmos supervisionados.

---

### Principais Tarefas do Aprendizado Não Supervisionado

#### 🔹 1. Clusterização (agrupamento)

> **Objetivo:** descobrir grupos de instâncias similares sem saber previamente a qual grupo elas pertencem.


- **Exemplos de aplicação:**  
  - Segmentação de clientes com base em comportamento de compra  
  - Agrupamento de documentos ou notícias por temas  
  - Identificação de padrões de tráfego em redes

#### 🔹 2. Redução de Dimensionalidade

> **Objetivo:** representar os dados em um espaço de menor dimensão, preservando o máximo de informação possível.

- **Exemplos de aplicação:**  
  - Visualização de dados de alta dimensão  
  - Compressão de dados  
  - Eliminação de redundâncias antes de aplicar algoritmos supervisionados

#### 🔹 3. Detecção de Anomalias

> **Objetivo:** identificar exemplos que se desviam do padrão geral dos dados — os chamados *outliers*.

- **Exemplos de aplicação:**  
  - Detecção de fraudes bancárias  
  - Diagnóstico de falhas em máquinas industriais  
  - Identificação de comportamentos incomuns em segurança cibernética

#### 🔹 4. Modelagem de Densidade

> **Objetivo:** estimar a função de distribuição de probabilidade dos dados.


- **Exemplos de aplicação:**  
  - Geração de dados sintéticos  
  - Filtragem bayesiana  
  - Avaliação de risco

#### 🔹 5. Aprendizado de Representações Latentes (Extração de características)

> **Objetivo:** aprender representações compactas e úteis dos dados, que possam ser reutilizadas em outras tarefas.

- **Exemplos de aplicação:**  
  - Compressão e reconstrução de imagens  
  - Representação vetorial de palavras em NLP  
  - Descoberta de temas ocultos em grandes corpora de texto
